
# Prima anual vs fraude

```python
df = pd.read_csv("PFDA_fraud_car.csv", sep=';')
```

```python
print("📋 Lista de columnas del dataset:")
print(df.columns.tolist())
```

```python
print("📋 Columnas del dataset:\n")
for col in df.columns:
    print("-", col)
```

```python
columna_prima = "Annual_Premium"
```

```python
if columna_prima in df.columns:
    grafico = sns.boxplot(x='FraudFound_P', y=columna_prima, data=df)
    grafico.set_title("Distribución de la Prima Anual según fraude detectado")
    grafico.set_xlabel("Fraude detectado (0 = No, 1 = Sí)")
    grafico.set_ylabel("Prima Anual")
    print("\n✅ Gráfico generado correctamente")
    print("\nPromedio de prima anual según fraude:")
    print(df.groupby('FraudFound_P')[columna_prima].mean())
else:
    print(f"⚠️ La columna '{columna_prima}' no existe. Aquí tienes las disponibles:")
    print(df.columns.tolist())
```
